```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ave_income=54200
income_arr=np.array([35500,41800,45200,49600,66600,80200])
education_arr = np.arange(len(income_arr))
```

```python
plt.plot(education_arr,income_arr,"o-",label='Relatiionship between income and education')
plt.xlabel('Education level')
plt.ylabel('Income')

# Create an array for average income with the same length as education_arr
ave_income_arr = np.full(len(education_arr), ave_income)
plt.plot(education_arr,ave_income_arr,label='Average income') # Plot using the new array

plt.legend()

plt.show()
```

```python
education_levels = np.array([
"Less than high school completion",
"High school completion",
"Some college, no degree",
"Associate’s degree",
"Bachelor’s degree",
"Master’s or higher degree"
])
```

```python
plt.plot(education_levels,income_arr,"o-",label='Relatiionship between income and education')
plt.xlabel('Education level')
plt.ylabel('Income')

plt.xticks(education_arr, education_levels, rotation=45, ha="right")
ave_income_arr = np.full(len(education_levels), ave_income)
plt.plot(education_levels,ave_income_arr,label='Average income')

plt.legend()

plt.show()
```

```python
z_linear = np.polyfit(education_arr,income_arr,1)
print("Slope = ", z_linear[0], " in dollar per level")
print("Y-intercept = ", z_linear[1], " in dollar at level 0, i.e., Less than high school completion")
```
